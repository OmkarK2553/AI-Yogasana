```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import mediapipe as mp
import cv2 as cv
import time as time
```

```python
mp_pose = mp.solutions.pose
mp_drawing = mp.solutions.drawing_utils
pose0 = mp_pose.Pose(static_image_mode=True,min_detection_confidence=0.3, model_complexity=0)
pose1 = mp_pose.Pose(static_image_mode=True,min_detection_confidence=0.3, model_complexity=1)
pose2 = mp_pose.Pose(static_image_mode=True,min_detection_confidence=0.3, model_complexity=2)
```

```python
landmarks_list_pose0 = []
landmarks_list_pose1 = []
landmarks_list_pose2 = []
```

```python
img1 = cv.imread(r"D:\OpenCV Practice\Comparative_Study\images\4.png")
```

```python
start_time = time.time()
results1_pose0 = pose0.process(cv.cvtColor(img1,cv.COLOR_BGR2RGB))
end_time = time.time()
latency1_pose0 = end_time - start_time

start_time = time.time()
results1_pose1 = pose1.process(cv.cvtColor(img1,cv.COLOR_BGR2RGB))
end_time = time.time()
latency1_pose1 = end_time - start_time

start_time = time.time()
results1_pose2 = pose2.process(cv.cvtColor(img1,cv.COLOR_BGR2RGB))
end_time = time.time()
latency1_pose2 = end_time - start_time

print(latency1_pose0)
print(latency1_pose1)
print(latency1_pose2)
```

```python
img1_copy = img1.copy()

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

if results1_pose0.pose_landmarks:
    img1_copy = img1.copy()
    mp_drawing.draw_landmarks(image=img1_copy,landmark_list=results1_pose0.pose_landmarks,connections=mp_pose.POSE_CONNECTIONS)
    axes[0].imshow(img1_copy[:,:,::-1])
    axes[0].set_title("Output image-0")
    axes[0].axis("Off")

if results1_pose1.pose_landmarks:
    img1_copy = img1.copy()
    mp_drawing.draw_landmarks(image=img1_copy,landmark_list=results1_pose1.pose_landmarks,connections=mp_pose.POSE_CONNECTIONS)
    axes[1].imshow(img1_copy[:,:,::-1])
    axes[1].set_title("Output image-1")
    axes[1].axis("Off")

if results1_pose2.pose_landmarks:
    img1_copy = img1.copy()
    mp_drawing.draw_landmarks(image=img1_copy,landmark_list=results1_pose2.pose_landmarks,connections=mp_pose.POSE_CONNECTIONS)
    axes[2].imshow(img1_copy[:,:,::-1])
    axes[2].set_title("Output image-2")
    axes[2].axis("Off")

plt.tight_layout()
plt.show()
```

```python
curr_list = []
for i in range(33):
    curr_list.append(f'{mp_pose.PoseLandmark(i).name}:\n{results1_pose0.pose_landmarks.landmark[mp_pose.PoseLandmark(i).value]}')
landmarks_list_pose0.append(curr_list)
curr_list = []
for i in range(33):
    curr_list.append(f'{mp_pose.PoseLandmark(i).name}:\n{results1_pose1.pose_landmarks.landmark[mp_pose.PoseLandmark(i).value]}')
landmarks_list_pose1.append(curr_list)
curr_list = []
for i in range(33):
    curr_list.append(f'{mp_pose.PoseLandmark(i).name}:\n{results1_pose2.pose_landmarks.landmark[mp_pose.PoseLandmark(i).value]}')
landmarks_list_pose2.append(curr_list)
```
